```python
team_counts = df['Team'].value_counts()
print("\nNumber of employees in each team:")
print(team_counts)

#  Calculate the Percentage Split for Each Team
total_employees = len(df)
team_percentage = (team_counts / total_employees) * 100
print("\nPercentage split for each team:")
print(team_percentage)

#  Combine Counts and Percentages
team_distribution = pd.DataFrame({'Count': team_counts, 'Percentage': team_percentage})
print("\nCombined team distribution (Count and Percentage):")
print(team_distribution)
```

```python
positions_group = df.groupby('Position')

# Display the groups
for position, group in positions_group:
    print(f"Position: {position}")
    print(group)
    print("\n")
```

```python
# Step 3: Define Age Groups
bins = [20, 24, 29, 34, 39, 44, 49, 54, 59, 64, 69, 74, 79, 84, 89, 94, 99]
labels = ['20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75-79', '80-84', '85-89', '90-94', '95-99']

# Assign each employee to an age group
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Display the first few rows of the dataset with the new 'Age Group' column
print("Dataset with Age Group column:")
print(df.head())

# Step 4: Count the Number of Employees in Each Age Group
age_group_counts = df['Age Group'].value_counts().sort_index()

# Display the counts
print("\nNumber of employees in each age group:")
print(age_group_counts)

# Step 5: Identify the Predominant Age Group
predominant_age_group = age_group_counts.idxmax()
predominant_age_group_count = age_group_counts.max()

# Display the predominant age group
print(f"\nThe predominant age group is {predominant_age_group} with {predominant_age_group_count} employees.")
```

```python
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

# Display the result
print("\nSalary expenditure by team and position:")
print(salary_expenditure)

# Step 4: Identify the Team and Position with the Highest Salary Expenditure
highest_salary_expenditure = salary_expenditure.loc[salary_expenditure['Salary'].idxmax()]

# Display the result
print(f"\nThe team and position with the highest salary expenditure is {highest_salary_expenditure['Team']} - {highest_salary_expenditure['Position']} with a total salary of {highest_salary_expenditure['Salary']}.")
```

```python
#  Calculate the Correlation Coefficient
correlation = df[['Age', 'Salary']].corr()

# Display the correlation coefficient
print("\nCorrelation coefficient between Age and Salary:")
print(correlation)

#  Visualize the Data with a Scatter Plot
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Create a scatter plot of age vs. salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)

# Add title and labels
plt.title('Scatter Plot of Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')

# Show the plot
plt.show()

# Step 5: Add a Regression Line to the Scatter Plot
# Create a scatter plot with a regression line of age vs. salary
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Salary', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})

# Add title and labels
plt.title('Scatter Plot of Age vs. Salary with Regression Line')
plt.xlabel('Age')
plt.ylabel('Salary')

# Show the plot
plt.show()
```
